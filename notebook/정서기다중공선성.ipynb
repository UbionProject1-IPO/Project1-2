## 상관계수확인

```python
# 상관계수 분포 파악하기

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

# 1) 상관계수 행렬 계산
corr = multi.corr().abs()  # 절대값으로

# 2) upper triangle만 추출 (중복 제거)
corr_vals = corr.where(~np.eye(corr.shape[0],dtype=bool))
corr_vals = corr_vals.unstack().dropna()

# 3) 히스토그램으로 분포 확인
plt.figure(figsize=(8,4))
plt.hist(corr_vals, bins=50)
plt.xlabel('절대 상관계수')
plt.ylabel('빈도')
plt.title('변수쌍별 상관계수 분포')
plt.show()
```

```python
#높은 상관쌍(Threshold)만 추출하기

import pandas as pd

# 1) 상관계수 행렬 계산
corr = multi.corr().abs()

# 2) 상삼각행렬 mask
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)

# 3) 임계값 이상의 쌍 필터링
threshold = 0.6
high_corr = (
    corr.where(mask)
        .stack()
        .reset_index()
        .rename(columns={'level_0':'var1','level_1':'var2',0:'corr'})
)
high_corr = high_corr[high_corr['corr'] > threshold].sort_values('corr', ascending=False)

print(f"총 {len(high_corr)}개 쌍이 |r| > {threshold}")
high_corr.head(57)
```

```python
# 계층적 군집 (Clustermap)으로 히트맵 그리기

import seaborn as sns
import matplotlib.pyplot as plt

# 변수 표준화 (선택)
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(multi)

# clustermap 그리기
sns.clustermap(
    pd.DataFrame(X_std, columns=multi.columns).corr(),
    method='average',      # linkage 방법
    metric='euclidean',    # 거리 측정
    cmap='vlag',
    figsize=(12,12),
    vmax=1, vmin=-1,
    linewidths=.5
)
plt.suptitle('Hierarchical Clustering of Variable Correlations', y=1.02)
plt.show()
```

```python
# PCA 로딩(Loading) 분석으로 군집 확인

import numpy as np
from sklearn.decomposition import PCA

# 1) 표준화
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(multi)

# 2) PCA 수행
pca = PCA(n_components=10)   # 상위 10개 주성분
pca.fit(X_std)

# 3) 로딩 행렬 (주성분 × 변수)
loadings = pd.DataFrame(
    pca.components_.T,
    index=multi.columns,
    columns=[f'PC{i+1}' for i in range(10)]
)

# 4) 특정 PC에서 절댓값 큰 상위 변수 보기
for pc in loadings.columns:
    top_vars = loadings[pc].abs().sort_values(ascending=False).head(5).index.tolist()
    print(f"{pc} 상위 5개 로딩 변수: {top_vars}")
```

## PCA 비지도학습

```python
sm = pd.read_csv('../data/주가수익률/주가수익률_동행순환변동치_smooth.csv')
sm = sm[['회사명','경기국면']]
sm.info()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt

# 1) 데이터 불러오기 및 전처리
multi = pd.read_csv('../data/주가수익률/주가수익률_base.csv')
multi = multi.drop(columns=['상장일','stock_code','회사명']).dropna().reset_index(drop=True)

# 2) 표준화
X_std = StandardScaler().fit_transform(multi.values)

# 3) PCA – 설명 분산 80% 확보
pca_full = PCA().fit(X_std)
cum_var = np.cumsum(pca_full.explained_variance_ratio_)
n_comp = np.searchsorted(cum_var, 0.8) + 1
print(f"80% 설명분산 확보를 위한 PC 개수: {n_comp}")

pca = PCA(n_components=n_comp, random_state=42)
X_pca = pca.fit_transform(X_std)

# 4) 최적 k 탐색 (실루엣 스코어)
sil_scores = []
ks = list(range(2, 11))
for k in ks:
    km = KMeans(n_clusters=k, random_state=42).fit(X_pca)
    sil_scores.append(silhouette_score(X_pca, km.labels_))

plt.figure(figsize=(6,4))
plt.plot(ks, sil_scores, marker='o')
plt.xlabel('Number of clusters k')
plt.ylabel('Silhouette Score')
plt.title('KMeans Silhouette Analysis')
plt.grid(True)
plt.show()

best_k = ks[np.argmax(sil_scores)]
print(f"최적의 k (실루엣 기준): {best_k}")

# 5) 여러 알고리즘 비교
algos = {
    'KMeans':        KMeans(n_clusters=best_k, random_state=42).fit_predict(X_pca),
    'Agglomerative': AgglomerativeClustering(n_clusters=best_k).fit_predict(X_pca),
    'DBSCAN':        DBSCAN(eps=1.0, min_samples=5).fit_predict(X_pca),
    'GMM':           GaussianMixture(n_components=best_k, random_state=42).fit_predict(X_pca)
}

for name, labels in algos.items():
    uniq = set(labels)
    if len(uniq) > 1 and -1 not in uniq:
        score = silhouette_score(X_pca, labels)
        print(f"{name}: clusters={len(uniq)}, silhouette={score:.3f}")
    else:
        print(f"{name}: clusters={len(uniq)}")

# 6) 최종 클러스터링 (예: KMeans)
multi_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_comp)])
multi_pca['cluster'] = algos['KMeans']

# 7) 각 클러스터 프로파일링 (원본 변수 평균)
profile = multi.join(multi_pca['cluster']).groupby('cluster').mean()
print(profile)

# 8) 각 PC별 “주요 기여 변수” 확인
loadings = pd.DataFrame(
    pca.components_.T,
    index=multi.columns,
    columns=[f'PC{i+1}' for i in range(n_comp)]
)
pc_top_vars = {
    pc: loadings[pc].abs().sort_values(ascending=False).head(10).index.tolist()
    for pc in loadings.columns
}
for pc, vars_ in pc_top_vars.items():
    print(f"{pc} 상위 기여 변수: {vars_}")

# ───────────────────────────────────────────────────────
# 9) Random Forest로 cluster 예측 및 성능 평가
#    (원본 176개 변수 → cluster 레이블)

# 9.1) 데이터 준비
df_rf = multi.copy()
df_rf['cluster'] = multi_pca['cluster']

X = df_rf.drop(columns=['cluster'])
y = df_rf['cluster']

# 9.2) train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 9.3) RF 학습
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)

# 9.4) 피처 임포턴스 시각화
importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(8,6))
importances.head(20).plot.barh()
plt.gca().invert_yaxis()
plt.xlabel('Feature Importance')
plt.title('Top 20 Feature Importances (RF)')
plt.tight_layout()
plt.show()

print("Top 20 important features:")
print(importances.head(20))

# 9.5) 예측 및 혼동행렬 출력
y_pred = rf.predict(X_test)

# classification report
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

# confusion matrix 시각화
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
fig, ax = plt.subplots(figsize=(6,6))
disp.plot(ax=ax, cmap='Blues', values_format='d')
plt.title('Confusion Matrix')
plt.show()
```

## 다중공선성 제외 -> 클러스터링 -> RF

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. Load and preprocess multi data
multi = pd.read_csv('../data/주가수익률_raw/주가수익률_base.csv')
multi = multi.drop(columns=['상장일', 'stock_code']).dropna().reset_index(drop=True)

# 2. Check and remove multicollinearity via VIF
def calculate_vif(df, thresh=10.0):
    variables = df.select_dtypes(include=[np.number]).columns.tolist()
    while True:
        vif = pd.Series(
            [variance_inflation_factor(df[variables].values, i) for i in range(len(variables))],
            index=variables
        )
        max_vif = vif.max()
        if max_vif > thresh:
            drop_var = vif.idxmax()
            print(f"Dropping '{drop_var}' with VIF={max_vif:.2f}")
            variables.remove(drop_var)
        else:
            break
    return df[variables]

# Preserve original numeric feature list
X_numeric = multi.drop(columns=['회사명'])
original_features = X_numeric.columns.tolist()

# Perform VIF filtering
X_vif = calculate_vif(X_numeric, thresh=10.0)
print("Remaining features after VIF filtering:", X_vif.columns.tolist())

# Derive dropped_columns list
dropped_columns = [col for col in original_features if col not in X_vif.columns.tolist()]
print("Dropped columns due to high VIF:", dropped_columns)

# 3. Merge 경기국면 into multi
sm = pd.read_csv('../data/주가수익률/주가수익률_동행순환변동치-smooth.csv')
sm = sm[['회사명', '경기국면']]
multi = multi.merge(sm, on='회사명')

# 4. Clustering (include 경기국면 as independent variable)
le_phase = LabelEncoder()
multi['경기국면_encoded'] = le_phase.fit_transform(multi['경기국면'])

# Prepare clustering features
X_cluster = pd.concat([X_vif, multi[['경기국면_encoded']]], axis=1)

# Scale features
scaler = StandardScaler()
X_cluster_scaled = scaler.fit_transform(X_cluster)

# Silhouette analysis
range_n_clusters = range(2, 11)
silhouette_avgs = []
for n_clusters in range_n_clusters:
    km = KMeans(n_clusters=n_clusters, random_state=42)
    labels = km.fit_predict(X_cluster_scaled)
    score = silhouette_score(X_cluster_scaled, labels)
    silhouette_avgs.append(score)
    print(f"n_clusters={n_clusters}, silhouette score={score:.3f}")

plt.figure()
plt.plot(list(range_n_clusters), silhouette_avgs, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Average silhouette score')
plt.title('Silhouette Analysis')
plt.show()

# Determine best cluster count
best_n = range_n_clusters[np.argmax(silhouette_avgs)]
print(f"Best number of clusters: {best_n}")

# Detailed silhouette plot for best_n
km = KMeans(n_clusters=best_n, random_state=42)
cluster_labels = km.fit_predict(X_cluster_scaled)
sil_values = silhouette_samples(X_cluster_scaled, cluster_labels)

plt.figure(figsize=(8, 6))
plt.xlim([-0.1, 1])
plt.ylim([0, len(X_cluster_scaled) + (best_n + 1) * 10])
y_lower = 10
for i in range(best_n):
    ith_sil_values = np.sort(sil_values[cluster_labels == i])
    size_cluster = ith_sil_values.shape[0]
    y_upper = y_lower + size_cluster
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_sil_values)
    y_lower = y_upper + 10

plt.title(f"Silhouette Plot for {best_n} Clusters")
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster Label")
plt.show()

# Assign cluster labels
multi['cluster'] = cluster_labels

# 5. Random Forest to predict clusters
features = X_cluster.columns.tolist()
X = scaler.fit_transform(multi[features])
y = multi['cluster']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train RF
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

# Confusion matrix & classification report
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
target_names = [str(c) for c in sorted(np.unique(y))]
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=target_names))

# Feature importances
feat_importances = pd.Series(rf.feature_importances_, index=features)
feat_importances.index = feat_importances.index.str.replace('경기국면_encoded', '경기국면')
feat_importances = feat_importances.sort_values(ascending=False)
print("Feature Importances:\n", feat_importances)

# Sorted feature list
sorted_features = feat_importances.index.tolist()
print("Features sorted by importance:", sorted_features)

# Plot importances
plt.figure(figsize=(10, 6))
plt.barh(feat_importances.index, feat_importances.values)
plt.xlabel('Importance')
plt.title('Random Forest Feature Importances for Cluster Prediction')
plt.gca().invert_yaxis()
plt.show()
```
