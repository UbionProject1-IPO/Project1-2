```python
dataset = pd.read_csv(r'..\data\경기국면_raw\경기종합지수.csv',encoding='utf-8')
dataset
```

```python
data = dataset[['시점',' 동행종합지수(2020=100)']]
data
```

```python
import pandas as pd

# 1) 복사본 만들기 ― SettingWithCopyWarning 피하기
data = dataset[['시점', ' 동행종합지수(2020=100)']].copy()

# 2) 문자열 정리
data['시점'] = (
    data['시점']        # 시점 컬럼
    .astype(str)        # 혹시 숫자형이어도 문자열로 변환
    .str.strip()        # 앞뒤 공백 제거
    .str.replace('.', '-', regex=False)  # '2010.01' → '2010-01'
)

# 3) 연도만 있는 경우(길이 4) → ‘YYYY-01’로 보강
mask_year_only = data['시점'].str.len() == 4
data.loc[mask_year_only, '시점'] = data.loc[mask_year_only, '시점'] + '-10'

# 4) 날짜형으로 변환 후 ‘YYYY-MM’ 문자열로 통일
data['시점'] = pd.to_datetime(data['시점'], format='%Y-%m')
data['시점'] = data['시점'].dt.strftime('%Y-%m')

# 확인
print(data.head())
```
