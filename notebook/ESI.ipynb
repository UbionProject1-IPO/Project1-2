```python
import pandas as pd
```

```python
df = pd.read_csv('../data/ESI_raw.csv', encoding='cp949')

df.head()
```

```python
import pandas as pd


# 2. '계정코드별' 컬럼을 인덱스로 설정
df.set_index('계정코드별', inplace=True)

# 3. 전치(transpose)하여 날짜를 행으로 만들기
df = df.T

# 4. 인덱스 이름 변경 (선택)
df.index.name = '날짜'

# 5. 숫자형으로 변환 (필요시)
df = df.apply(pd.to_numeric, errors='coerce')

```

```python

df.head()
```

```python
# 4. 날짜 포맷 변경 (예: '2024.12' → datetime 형태)
df.index = pd.to_datetime(df.index, format='%Y.%m')

# 5. 지수 컬럼 숫자형으로 변환
df = df.apply(pd.to_numeric, errors='coerce')
```
